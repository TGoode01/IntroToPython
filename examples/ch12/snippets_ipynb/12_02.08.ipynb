## 17.2.8 Inflection: Pluralization and Singularization

```python
from textblob import Word
```

```python
index = Word('index')
```

```python
index.pluralize()
```

```python
cacti = Word('cacti')
```

```python
cacti.singularize()
```

```python
from textblob import TextBlob
```

```python
animals = TextBlob('dog cat fish bird').words
```

```python
animals.pluralize()
```
